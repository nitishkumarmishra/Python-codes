# Differentially expressed genes and proteins in Unt VS TGFb

## Gene expression :: DEG

### Nitish RNAseq data

```r
DEG_NK_path <- "C:/Users/nmishra/Dropbox/PC/Desktop/Total RNAseq EMT/jupyter codes/out/limma-voom.mrna"
dir_RNAseq_path <- "170224.rdna_rn18s/NMuMG"
file <- sprintf("%s/%s/blancgrp_211613_RNAseq_total_stranded.unt--vs--tgfb.diff-all.txt",DEG_NK_path, dir_RNAseq_path)
file
```

```r
DEG.Unt_VS_TGFb_NK.transcription <- read.table(file=file, 
                header=TRUE, sep="\t", row.names=1,
                quote="", comment.char="#", stringsAsFactors=F)
head(DEG.Unt_VS_TGFb_NK.transcription)
dim(DEG.Unt_VS_TGFb_NK.transcription)
```

### Hyunsoo RNAseq data

```r
DEG_HK_path <- "C:/Users/nmishra/Dropbox/PC/Desktop/HK EXIT STRATEGY/jupyter codes/out/limma-voom.mrna"
dir_RNAseq_path <- "170224.rdna_rn18s/NMuMG"
file <- sprintf("%s/%s/170224.NMuMG.limma-voom.mrna.unt48--vs--tgfbCX5461100nm.diff-all.txt",DEG_HK_path, dir_RNAseq_path)
```

```r
DEG.Unt_VS_TGFb_HK.transcription <- read.table(file=file, 
                header=TRUE, sep="\t", row.names=1,
                quote="", comment.char="#", stringsAsFactors=F)
head(DEG.Unt_VS_TGFb_HK.transcription)
dim(DEG.Unt_VS_TGFb_HK.transcription)
```

### Venn diagram of common genes

```r
length(intersect(rownames(DEG.Unt_VS_TGFb_NK.transcription), rownames(DEG.Unt_VS_TGFb_HK.transcription)))
```

```r
file_name <- "DEG.Unt_VS_TGFb_NKM_VS_HK.transcription"
ven <- venndetail(list(NKM = rownames(DEG.Unt_VS_TGFb_NK.transcription), HK = rownames(DEG.Unt_VS_TGFb_HK.transcription)))
plot(ven, main="Common gene")
print("DEG in Unt VS TGFb in NKM and HK analysis")
tiff(file=sprintf("./tiff/Venn Diagram .%s.tiff", file_name), width=6, height=6, units="in", res=300)
plot(ven)
dev.off()
```
